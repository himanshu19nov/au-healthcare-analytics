```python
import pandas as pd

elective_file = "../data/raw/Elective_Filtered.xlsx"

elective_data = pd.read_excel(
    elective_file,
    sheet_name=None
)
```

```python
elective_data.keys()
```

```python
t24 = elective_data["Table 2.4"].copy()
t31 = elective_data["Table 3.1"].copy()
t42 = elective_data["Table 4.2"].copy()
t410 = elective_data["Table 4.10"].copy()
```

```python
t24.tail(10)
```

```python
t24.head(10)


```

```python
t24["Measure"].unique()
```

```python
t24["State/territory"].unique()
```

```python
t24_clean = t24.copy()

# Remove rows where measure is missing
t24_clean = t24_clean[t24_clean["Measure"].notna()]

# Remove Total rows
t24_clean = t24_clean[t24_clean["State/territory"] != "Total"]

# Drop derived columns
t24_clean = t24_clean.drop(columns=[
    "Average annual change (%) last five years(a)",
    "Change (%) since last year(a)"
])

t24_clean.head(10)
```

```python
t24_clean["State/territory"].unique()
```

```python
t24_clean["Measure"].unique()
```

```python
t24_clean.head()
```

```python
t24_clean.shape
```

```python
years = ["2020–21","2021–22","2022–23","2023–24","2024–25(b)"]

t24_long = t24_clean.melt(
    id_vars=["State/territory","Measure"],
    value_vars=years,
    var_name="year",
    value_name="value"
)

t24_long.head()
```

```python
t24_long.shape
```

```python
t24_long = t24_long.rename(columns={
    "State/territory": "state",
    "Measure": "metric"
})
```

```python
t24_long.info()
```

```python
t24_long["metric"].value_counts()
```

```python
t24_long["metric"] = t24_long["metric"].str.replace("(b)", "", regex=False)
```

```python
t24_long["metric"].unique()
```

```python
t24_long = t24_long.sort_values(["state","metric","year"])
```

```python
t24_long = t24_long.reset_index(drop=True)
```

```python
t24_long.info()
```

```python
t24_long.to_csv("elective_admissions.csv", index=False)
```

```python
t31_clean = t31.copy()

# remove rows with missing values (footnotes)
t31_clean = t31_clean.dropna(subset=["2020–21"])

# remove total row
t31_clean = t31_clean[~t31_clean["Surgical specialty"].str.contains("Total", na=False)]

# drop derived columns
t31_clean = t31_clean.drop(columns=[
    "Average annual change (%) last five years(a)",
    "Change (%) since last year(a)"
])

t31_clean.head()
```

```python
t31_clean.shape
```

```python
t31_clean.head()
```

```python
years = ["2020–21","2021–22","2022–23","2023–24","2024–25(b)"]

t31_long = t31_clean.melt(
    id_vars=["Surgical specialty"],
    value_vars=years,
    var_name="year",
    value_name="value"
)

t31_long.head()
```

```python
t31_long = t31_long.rename(columns={
    "Surgical specialty": "surgical_specialty"
})
```

```python
t31_long["year"] = t31_long["year"].str.replace("(b)", "", regex=False)
```

```python
t31_long["surgical_specialty"] = t31_long["surgical_specialty"].str.replace("(b)", "", regex=False)
```

```python
t31_long.shape
```

```python
t31_long.head()
```

```python
t31_long.to_csv("elective_surgical_specialty.csv", index=False)
```

```python
t410_clean = t410.copy()

# remove total row
t410_clean = t410_clean[t410_clean["Clinical urgency"] != "Total"]

# drop total column
t410_clean = t410_clean.drop(columns=["Total"])

t410_clean.head()
```

```python
t410_clean = t410_clean[
    t410_clean["Clinical urgency"].isin([
        "Category 1",
        "Category 2",
        "Category 3"
    ])
]
```

```python
t410_clean["Clinical urgency"].unique()
```

```python
t410_clean.shape
```

```python
states = ["NSW","Vic","Qld","WA","SA","Tas","ACT","NT"]

t410_long = t410_clean.melt(
    id_vars=["Clinical urgency","Measure"],
    value_vars=states,
    var_name="state",
    value_name="value"
)
```
