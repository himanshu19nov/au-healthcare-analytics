# AU Health Analytics - EDA
This notebook imports the master tables from the three AIHW Excel files as defined in `Tables.txt`.

```python
import pandas as pd
import os

# File paths
ed_file = 'Emergency-department-care-2024-25.xlsx'
elective_file = 'Elective-surgery-waiting-times-2024-25.xlsx'
expenditure_file = 'hwe-105-health-expenditure-australia-2023-24-supplementary-tables.xlsx'

# Tables to import (from Tables.txt)
ed_tables = ['Table 2.2', 'Table 3.1', 'Table 4.5', 'Table 4.9', 'Table 5.1', 'Table 5.3', 'Table 6.1', 'Table 6.3']
elective_tables = ['Table 2.1', 'Table 2.4', 'Table 3.1', 'Table 3.2', 'Table 4.1', 'Table 4.2', 'Table 4.4', 'Table 4.10']
expenditure_tables = ['Table 3', 'Table 4', 'Table 5', 'Table 28', 'Table 29', 'Table 30', 'Table A3', 'Table B3', 'Table B6', 'Table B9', 'Table B12', 'Table B15', 'Table B18', 'Table B21', 'Table B24']
```

```python
def load_tables(file_path, tables):
    data = {}
    print(f"Loading tables from {file_path}...")
    for table in tables:
             # header=1 skips the title row and uses the second row as columns
        df = pd.read_excel(file_path, sheet_name=table, header=1)
    
             # Remove rows that are entirely empty (common in Excel footers)
        df = df.dropna(how='all').reset_index(drop=True)
   
        data[table] = df
        print(f"  Loaded {table}")
    return data
   
   #Load data into dictionaries
ed_data = load_tables(ed_file, ed_tables)
elective_data = load_tables(elective_file, elective_tables)
expenditure_data = load_tables(expenditure_file, expenditure_tables)
```
